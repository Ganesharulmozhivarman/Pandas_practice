# Melt Function - used to convert wide format to long format 

```python
# Read Data Set 
life_df = pd.read_csv('life_expectancy_years.csv')
```

```python
#head 
life_df.head()
```

```python
#shape
life_df.shape
```

```python
#melt Function 
df_melted=pd.melt(life_df,id_vars='country',var_name='Year',value_name='Life Expectancy')
df_melted
```

# Concat Function 


```python
import pandas as pd
```

```python
# dataframes having the same colums
df1 = pd.DataFrame({'Name': ['Arijit', 'Neeraj', 'Sakshi', 'Muskan'],
'age': [34, 30, 22, 33],
'Gender': ['M', 'MY', 'FY', 'H']}

)

df2 = pd.DataFrame({'Name': ['kartik', "Veer", 'Preeti'],
'age': [31,22, 19],
'Gender': ['M', 'M', 'F']}

)

 
```

```python
# concating two data frames (union in sql)

df=pd.concat([df1,df2],axis=0) # 0 -row wise concat
df
```

```python
# resetting the old index and creating a new index
df=df.reset_index()
df
```

```python
#Removing the index xolm
df=df.drop(['index'],axis=1)
df
```

# Merge Function (similar to joins in sql)

```python
# creating Ist data frame
dff1 = pd.DataFrame({'pincode' :[249201,175131,263136,246171,171004],
'city_name':['Rishikesh' , 'Manali', 'shimtal', 'Rudraprayag' ,'shimla'],
'distance from_delhi' :[230,543,321,370,343]})
# creating 2nd datafrane
dff2 = pd.DataFrame({ 'pincode' :[249201,175131, 263001],
'district' :['Dehradun' , 'kullu', 'Nainital'],
'state' :[ 'Uttarakhand' , 'Himachal Pradesh' , 'Uttarakhand']})


dff2
```

```python
dff1
```

```python
# left join 
pd.merge(dff1,dff2,how='left',on='pincode')
```

```python
# Right join 
pd.merge(dff1,dff2,how='right',on='pincode')
```

```python
# inner join 
pd.merge(dff1,dff2,how='inner',on='pincode')
```
